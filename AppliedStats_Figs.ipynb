```python
import pandas as pd
import matplotlib.pyplot as plt

# load the dataset
df = pd.read_csv('Survey on Mess Food Usage by IITH Students.csv')

# extract the required columns
df = df[['You are?', 'Which meal do you usually skip in a day?']]

# group the data by sleep time and meal skipped
grouped = df.groupby(['You are?', 'Which meal do you usually skip in a day?']).size()

# create a multi-index dataframe to store the grouped data
grouped = grouped.unstack()

# plot the segmented bar charts
grouped.plot(kind='bar', stacked=True)

# set the chart title and axis labels
plt.title('Meal Skipping by Sleep Time')
plt.xlabel('Sleep Time')
plt.ylabel('Number of Responses')

# show the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# load the dataset
df = pd.read_csv('Survey on Mess Food Usage by IITH Students.csv')

# extract the required columns
df = df[['Family Income (Per Annum)', 'How often do you skip mess eating mess food?']]

# create a bar chart to display the counts of each income range
income_counts = df['Family Income (Per Annum)'].value_counts()
income_counts.plot(kind='bar')

# set the chart title and axis labels
plt.title('Family Income')
plt.xlabel('Income Range')
plt.ylabel('Number of Responses')

# show the chart
plt.show()

# group the data by income range and how often they skip mess food
grouped = df.groupby(['Family Income (Per Annum)', 'How often do you skip mess eating mess food?']).size()

# create a multi-index dataframe to store the grouped data
grouped = grouped.unstack()

# plot the stacked bar chart
grouped.plot(kind='bar', stacked=True)

# set the chart title and axis labels
plt.title('Skipping Mess Food by Family Income and Frequency')
plt.xlabel('Family Income')
plt.ylabel('Number of Responses')

# show the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# load the dataset
df = pd.read_csv('Survey on Mess Food Usage by IITH Students.csv')

# extract the required columns
df = df[['What degree are you pursuing in IITH?', 'What are probable reason(s) for you to skip the mess?']]

# group the data by degree and reason for skipping meals
grouped = df.groupby(['What degree are you pursuing in IITH?', 'What are probable reason(s) for you to skip the mess?']).size()

# create a multi-index dataframe to store the grouped data
grouped = grouped.unstack()

# plot the stacked bar chart
grouped.plot(kind='bar', stacked=True)

# set the chart title and axis labels
plt.title('Degree and Reasons for Skipping Meals')
plt.xlabel('Degree')
plt.ylabel('Number of Responses')

# show the chart
plt.show()
```
